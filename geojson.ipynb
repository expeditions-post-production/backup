```python
places_list = places_text.splitlines()
```

```python
import re
```

```python
coords_text = coords_text.replace(' ','').splitlines()
```

```python
coord_list = []
for coord in coords_text:
    if re.match(r'\d+\.\d+',coord) == None:
        if re.match(r'\d+°\d+′(?:\d+″)?((с\.ш\.|N)|[,?])\d+°\d+′(?:\d+″)?',coord) == None:
            print(coord)
            long = input('longitude: ')
            lat = input('latitude: ')
            coord_list.append([lat,long])
        else:
            grad_long, min_long, sec_long, grad_lat, min_lat, sec_lat = re.match(r'(\d+)°(\d+)′(?:(\d+)″)?(?:(?:с\.ш\.|N)|[,?])(\d+)°(\d+)′(?:(\d+)″)?',coord).groups()
            if sec_long == None:
                sec_long = 0
            if sec_lat == None:
                sec_lat = 0
            long = int(grad_long) + int(min_long) / 60 + int(sec_long) / 3600
            lat = int(grad_lat) + int(min_lat) / 60 + int(sec_lat) / 3600
            coord_list.append([lat,long])
    else:
        long, lat = coord.split(',')
        coord_list.append([lat,long])
```

```python
geojson = {
  "type": "FeatureCollection",
  "features": []
}
```

```python
no_dub = []
for place, coord in zip(places_list, coord_list):
    if (place, coord) not in no_dub:
        no_dub.append((place,coord))
        feature = {"type": "Feature", "geometry": {"type": "Point", "coordinates": coord}, "properties": {"name": place}}
        geojson['features'].append(feature)
```
